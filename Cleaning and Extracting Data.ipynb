### Finding number of people waiting in ED Waiting Room Variables

```
def waiting_before(a_start, a_end, a_ctas, b_start, b_end, b_ctas, same_ctas):
  check = bool(b_start < a_start and b_end > a_start)
  if (same_ctas):
    return (a_ctas == b_ctas and check)
  else:
    return check


def waiting_after(a_start, a_end, a_ctas, b_start, b_end, b_ctas, same_ctas):
  check = bool(b_start > a_start and b_end > a_end)
  if (same_ctas):
    return (a_ctas == b_ctas and check)
  else:
    return check

def in_between(a_start, a_end, a_ctas, b_start, b_end, b_ctas, same_ctas):
  check = bool(b_start > a_start and b_end < a_end)
  if (same_ctas):
    return (a_ctas == b_ctas and check)
  else:
    return check

def date_list_within_range(i, cut):
  if ((i - cut) <= 0):
    start = 0
  else:
    start = i - cut
  
  
  if ((i+cut) >= sample_n):
    end = sample_n
  else:
    end = i + cut
  return start, end


# general function
def find_num_of_patients(patient_num, col_start, col_end, start_index, end_index, check_function, same_ctas=False):
  count = 0
  patient_a_ctas = ctas[patient_num]
  patient_a_start = col_start[patient_num]
  patient_a_end = col_end[patient_num]
  for i in range(start_index, end_index):
    patient_b_ctas = ctas[i]
    patient_b_start = col_start[i]
    patient_b_end = col_end[i]
    if (check_function(patient_a_start, patient_a_end, patient_a_ctas, 
                       patient_b_start, patient_b_end, patient_b_ctas, 
                       same_ctas)):
        count += 1
  return count
```

```
Number_of_Patients_In_Waiting_Before = [0]*len(df)
Number_of_Patients_In_Waiting_Before_CTAS = [0]*len(df)

for i in range(0,len(df)):
  start, end = date_list_within_range(i, 50)
  Number_of_Patients_In_Waiting_Before[i] = find_num_of_patients(i, df['FIRST_CONTACT_DT'][start:end], df['INIT_TREAT_LOC_DT'][start:end], start,end, waiting_before, False)
  Number_of_Patients_In_Waiting_Before_CTAS[i] = find_num_of_patients(i, df['FIRST_CONTACT_DT'][start:end], df['INIT_TREAT_LOC_DT'][start:end], start,end, waiting_before, True)
```

## Finding Number of Patients Waiting for Discharge Upon Arrival

```
def waiting_for_dispatch(b_num, a_arrival, b_start, b_end,):
  status = disp_group[b_num]
  check = bool(status == 'Discharged' and 
              b_start < a_arrival and b_end > a_arrival)
  return check  

def find_num_of_patients_2(patient_num, col_start, col_end, start_index, end_index, check_function):
  count1 = 0
  count2 = 0
  patient_a_start = arrival[patient_num]
  patient_a_ctas = ctas[patient_num]
  for i in range(start_index, end_index):
    patient_b_ctas = ctas[i]
    patient_b_start = col_start[i]
    patient_b_end = col_end[i]
    if (pd.isnull(patient_b_start) == False and pd.isnull(patient_b_end) == False):
      if (check_function(i, patient_a_start, patient_b_start, patient_b_end)):
          count1 += 1
          if (patient_a_ctas == patient_b_ctas):
            count2 += 1
  return count1, count2

Number_of_Patients_Waiting_For_Discharge = [0]*len(df)
Number_of_Patients_Waiting_For_Discharge_CTAS = [0]*len(df)

for i in range(0, len(df)):
  start, end = date_list_within_range(i, 300)
  Number_of_Patients_Waiting_For_Discharge[i], Number_of_Patients_Waiting_For_Discharge_CTAS[i] = find_num_of_patients_2(i, df['PHYS_ASSESSMENT_DT'][start:end], df['LAST_CONTACT_DT'][start:end], start, end, waiting_for_dispatch)
  print(i)
```

## Calculating Total Time In Between Time Blocks and Cumulative

```
def find_total_time(patient_num, col_start, col_end):
  return (col_end[patient_num] - col_start[patient_num]).total_seconds() /60

def find_conditional_total_time(patient_num, col_start, col_end_1, col_end_2):
  if pd.isnull(col_end_1[patient_num]):
    return find_total_time(patient_num, col_start, col_end_2)
  else:
    return find_total_time(patient_num, col_start, col_end_1)
```

```
tot_time_init_loc = [0]*len(df)
tot_time_init_treat_loc = [0]*len(df)
tot_time_phys = [0]*len(df)
tot_time_consult = [0]*len(df)
tot_time_bed = [0]*len(df)
tot_time_last = [0]*len(df)

tot_time_init_treat_loc_c = [0]*len(df)
tot_time_phys_c = [0]*len(df)
tot_time_consult_c = [0]*len(df)
tot_time_bed_c = [0]*len(df)
tot_time_last_c = [0]*len(df)
tot_time_for_treatment_c = [0]*len(df)

for i in range(0, len(df)):
  tot_time_init_loc[i] = find_total_time(i, df['FIRST_CONTACT_DT'], df['INIT_LOC_DT'])
  tot_time_init_treat_loc[i] = find_total_time(i, df['INIT_LOC_DT'], df['INIT_TREAT_LOC_DT'])
  tot_time_phys[i] = find_total_time(i, df['INIT_TREAT_LOC_DT'], df['PHYS_ASSESSMENT_DT'])
  tot_time_consult[i] = find_total_time(i, df['PHYS_ASSESSMENT_DT'], df['CONSULT_REQUESTED_DT'])
  tot_time_bed[i] = find_total_time(i, df['CONSULT_REQUESTED_DT'], df['IP_BED_ASSIGNED_DT'])
  tot_time_last[i] = find_total_time(i, df['IP_BED_ASSIGNED_DT'], df['LAST_CONTACT_DT'])

  tot_time_init_treat_loc_c[i] = find_total_time(i, df['FIRST_CONTACT_DT'], df['INIT_TREAT_LOC_DT'])
  tot_time_phys_c[i] = find_total_time(i, df['FIRST_CONTACT_DT'], df['PHYS_ASSESSMENT_DT'])
  tot_time_consult_c[i] = find_total_time(i, df['FIRST_CONTACT_DT'], df['CONSULT_REQUESTED_DT'])
  tot_time_bed_c[i] = find_total_time(i, df['FIRST_CONTACT_DT'], df['IP_BED_ASSIGNED_DT'])
  tot_time_last_c[i] = find_total_time(i, df['FIRST_CONTACT_DT'], df['LAST_CONTACT_DT'])
  
  #This is total time of treatment using "Decision to admit DT" might replace. 
  tot_time_for_treatment_c[i] = find_conditional_total_time(i, df['PHYS_ASSESSMENT_DT'], df['DEC_TO_ADMIT_DT'], df['LAST_CONTACT_DT'])
```

```
def find_weekday(dt):
  wk = dt.weekday()
  if wk == 0:
    return "Monday"
  elif wk == 1:
    return "Tuesday"
  elif wk == 2:
    return "Wednesday"
  elif wk == 3:
    return "Thursday"
  elif wk == 4:
    return "Friday"
  elif wk == 5: 
    return "Saturday"
  else:
    return "Sunday"


def find_num_of_doctors_at_arrival(patient_A_FC, df_phys, wk):
  num_docs = 0
  if (wk== "Sunday"):
    for i in range(0,24):
      if patient_A_FC == df_phys['hour'][i] and df_phys['day'][i] == wk:
        num_docs = df_phys['Physicians'][i]
        break
  elif (wk==  "Monday"):
    for i in range(24,48):
      if patient_A_FC == df_phys['hour'][i] and df_phys['day'][i] == wk:
        num_docs = df_phys['Physicians'][i]
        break
  elif (wk==  "Tuesday"):
    for i in range(48,72):
      if patient_A_FC == df_phys['hour'][i] and df_phys['day'][i] == wk:
        num_docs = df_phys['Physicians'][i]
        break
  elif (wk ==  "Wednesday"):
    for i in range(72,96):
      if patient_A_FC == df_phys['hour'][i] and df_phys['day'][i] == wk:
        num_docs = df_phys['Physicians'][i]
        break
  elif (wk ==  "Thursday"):
    for i in range(96,120):
      if patient_A_FC == df_phys['hour'][i] and df_phys['day'][i] == wk:
        num_docs = df_phys['Physicians'][i]
        break
  elif (wk == "Friday"):
    for i in range(120,145):
      if patient_A_FC == df_phys['hour'][i] and df_phys['day'][i] == wk:
        num_docs = df_phys['Physicians'][i]
        break
  elif (wk ==  "Saturday"):
    for i in range(145,167):
      if patient_A_FC == df_phys['hour'][i] and df_phys['day'][i] == wk:
        num_docs = df_phys['Physicians'][i]
        break


  return num_docs
  
Number_of_Physicians_At_Arrival = [0]*len(df)
Number_of_Physicians_At_3Hours_After = [0]*len(df)
Number_of_Physicians_At_2Hours_After = [0]*len(df)
Number_of_Physicians_At_1Hours_After = [0]*len(df)
Number_of_Physicians_At_3Hours_Before = [0]*len(df)
Number_of_Physicians_At_2Hours_Before = [0]*len(df)
Number_of_Physicians_At_1Hours_Before = [0]*len(df)

for i in range(0, len(df)):
    Number_of_Physicians_At_Arrival[i] = find_num_of_doctors_at_arrival(df['FIRST_CONTACT_DT'][i].hour, df_phys, find_weekday(df['FIRST_CONTACT_DT'][i]))
    Number_of_Physicians_At_3Hours_Before[i] = find_num_of_doctors_at_arrival((df['FIRST_CONTACT_DT'][i].hour - 3)%24, df_phys, find_weekday(df['FIRST_CONTACT_DT'][i]))
    Number_of_Physicians_At_2Hours_Before[i] = find_num_of_doctors_at_arrival((df['FIRST_CONTACT_DT'][i].hour - 2)%24, df_phys,find_weekday(df['FIRST_CONTACT_DT'][i]))
    Number_of_Physicians_At_1Hours_Before[i] = find_num_of_doctors_at_arrival((df['FIRST_CONTACT_DT'][i].hour - 1)%24, df_phys,find_weekday(df['FIRST_CONTACT_DT'][i]))
    Number_of_Physicians_At_1Hours_After[i] = find_num_of_doctors_at_arrival((df['FIRST_CONTACT_DT'][i].hour + 1)%24, df_phys,find_weekday(df['FIRST_CONTACT_DT'][i]))
    Number_of_Physicians_At_2Hours_After[i] = find_num_of_doctors_at_arrival((df['FIRST_CONTACT_DT'][i].hour + 2)%24, df_phys,find_weekday(df['FIRST_CONTACT_DT'][i]))
    Number_of_Physicians_At_3Hours_After[i] = find_num_of_doctors_at_arrival((df['FIRST_CONTACT_DT'][i].hour + 3)%24, df_phys,find_weekday(df['FIRST_CONTACT_DT'][i]))
```

```
#Helper Function
def date_list_within_range_other(i,n):
  cut = 50
  if ((i - cut) <= 0):
    start = 0
  else:
    start = i - cut
  
  
  if ((i+cut) >= n):
    end = n
  else:
    end = i + cut
  return start, end
```

### Number of Labs and DIs per patient

```
#Helper Function
def make_dictionary_for_count(data):
  values = data['VISIT_ID'].value_counts().keys().tolist()
  counts = data['VISIT_ID'].value_counts().tolist()
  return dict(zip(values, counts))


labs_dict = make_dictionary_for_count(df_labs)
DI_dict = make_dictionary_for_count(df_DI)

def dict_lookup(patient_sequence, labs_dict):
  try: 
    res = labs_dict[patient_sequence]
  except KeyError: 
    res = 0
  return res


for i in range(0, len(df)):
  Number_of_labs_for_patient[i] = dict_lookup(df['DATA_SEQU'][i], labs_dict)
  Number_of_DI_for_patient[i] = dict_lookup(df['DATA_SEQU'][i], DI_dict)
```

```
def find_labs_DI_before_hour_arrival_time(patient_arrival_time, labs_DI, start, end):
  result = 0
  for i in range(start, end):
      diff = (patient_arrival_time - labs_DI['ORD_REQ_DT'][i]).total_seconds() /60
      if diff > 0 and diff < 60:
        result = result+1
  return result

def find_labs_DI_after_hour_arrival_time(patient_arrival_time, labs_DI, start, end):
  result = 0
  for i in range(start, end):
      diff = (patient_arrival_time - labs_DI['ORD_REQ_DT'][i]).total_seconds() /60
      if diff < 0 and diff > -60:
        result = result+1
  return result
```

```
for i in range(0, len(df)):
  start_l, end_l = date_list_within_range_other(i, len(df_labs))
  Labs_Ordered_Hour_Before_Arrival[i] = find_labs_DI_before_hour_arrival_time(df['FIRST_CONTACT_DT'][i], df_labs[start_l:end_l],start_l, end_l)
  Labs_Ordered_Hour_After_Arrival[i] = find_labs_DI_after_hour_arrival_time(df['FIRST_CONTACT_DT'][i], df_labs[start_l:end_l],start_l, end_l)

for i in range(0, len(df)):
  start_d, end_d = date_list_within_range_other(i, len(df_DI))
  DIs_Ordered_Hour_Before_Arrival[i] = find_labs_DI_before_hour_arrival_time(df['FIRST_CONTACT_DT'][i], df_DI[start_d:end_d],start_d, end_d)
  DIs_Ordered_Hour_After_Arrival[i] = find_labs_DI_after_hour_arrival_time(df['FIRST_CONTACT_DT'][i], df_DI[start_d:end_d],start_d, end_d)
```

```
#Helper Function
def create_dict_for_time_delta(data):
  grouped = data.groupby('VISIT_ID', as_index=False).sum()
  d = grouped.set_index('VISIT_ID').transpose().to_dict()
  return d


def dict_lookup_array(patient_seq, dict): 
  try:
    res = dict[patient_seq]['TIME_DELTA']
  except KeyError:
    res = 0
  return res

labs_with_time = create_dict_for_time_delta(df_labs)
di_with_time = create_dict_for_time_delta(df_DI)
```

## What time of day was the patient's first contact?

```
def time_of_day_category(patient_first_contact):
  patient_first_contact = patient_first_contact.time()
  midnight = time(0,0,0)
  morning_dawn = time(6,0,0)
  noon = time(12, 0, 0)
  evening = time(18,0,0)
  if patient_first_contact > midnight and patient_first_contact < morning_dawn:
    return "Morning Dusk"
  elif patient_first_contact < noon and patient_first_contact > morning_dawn:
    return "Morning Dusk"
  elif patient_first_contact < evening and patient_first_contact > noon:
    return "Midday"
  else: 
    return "Night/Evening"

time_of_day_of_arrival = [0]*len(df)

for i in range(0, len(df)):
    time_of_day_of_arrival[i] = time_of_day_category(df['FIRST_CONTACT_DT'][i])
```
